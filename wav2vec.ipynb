# MWE for Norwegian wav2vec

```python
import librosa
from transformers import Wav2Vec2ForCTC, Wav2Vec2ProcessorWithLM
```

```python
audio_folder = '/media/pers/storage/storage/NB_tale_3_audio_segments/'
```

```python
audiofile = 'p1_g11_f2_2_t-free_29850_38856.wav'
```

```python
import IPython.display as display
display.Audio(audiofile, autoplay=False)
```

```python
audio, rate = librosa.load(audiofile, sr = 16000)
```

```python
model_name = 'NbAiLab/nb-wav2vec2-1b-bokmaal'
```

```python
processor = Wav2Vec2ProcessorWithLM.from_pretrained(model_name)
model = Wav2Vec2ForCTC.from_pretrained(model_name)
```

```python
input_values = processor(audio, sampling_rate = rate, return_tensors = 'pt').input_values
logits = model(input_values).logits
```

```python
transcription = processor.batch_decode(logits.detach().numpy()).text
```
